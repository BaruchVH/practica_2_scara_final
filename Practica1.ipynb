## Introducción a ROS 2

ROS 2 (Robot Operating System 2) es un conjunto de herramientas, bibliotecas y convenciones de software diseñadas para facilitar el desarrollo de aplicaciones robóticas. A diferencia de un sistema operativo tradicional, ROS 2 proporciona una infraestructura de comunicación distribuida que permite la interacción entre sensores, actuadores, algoritmos de control y sistemas de percepción mediante nodos conectados a través de tópicos, servicios y acciones.

Una de las principales ventajas de ROS 2 es su arquitectura modular, la cual permite desarrollar sistemas complejos a partir de componentes independientes. Gracias a esta característica, ROS 2 se ha convertido en uno de los estándares más utilizados en investigación, educación e industria para el desarrollo de aplicaciones robóticas.

Dentro de ROS 2, los proyectos se organizan en workspaces y paquetes. Un workspace corresponde al entorno general de trabajo donde se encuentran todos los paquetes del proyecto, mientras que cada paquete agrupa funcionalidades específicas como controladores, descripciones de robots, sensores o algoritmos de navegación.

En la presente práctica se utilizó un workspace denominado `ws_sem26_2`, dentro del cual se encuentran los paquetes `p0_py`, `scara_description` y `scara_bringup`, encargados respectivamente de los nodos desarrollados en Python, la descripción del robot y la configuración de simulación y visualización.

## URDF

URDF (Unified Robot Description Format) es un formato basado en XML utilizado para describir la estructura física de un robot. Este formato permite definir los enlaces que componen un manipulador, las articulaciones que los conectan y las propiedades geométricas, dinámicas e inerciales de cada componente.

Dentro de un archivo URDF se especifican elementos fundamentales como:

- Links (enlaces).
- Joints (articulaciones).
- Propiedades visuales.
- Propiedades de colisión.
- Propiedades inerciales.

Los links representan los cuerpos rígidos que conforman el robot, mientras que los joints describen la relación cinemática existente entre dos enlaces consecutivos. Gracias a esta estructura es posible representar cadenas cinemáticas completas y realizar simulaciones o visualizaciones en herramientas compatibles con ROS 2.

URDF constituye uno de los formatos más importantes dentro del ecosistema ROS debido a que sirve como base para la visualización, simulación y control de sistemas robóticos.

## Xacro

Xacro (XML Macros) es una extensión de URDF que permite reutilizar código mediante macros, propiedades y parámetros. Su principal objetivo es simplificar la construcción de modelos robóticos complejos evitando la repetición innecesaria de información.

Mientras que un archivo URDF tradicional puede volverse extenso y difícil de mantener, Xacro permite dividir la descripción del robot en múltiples archivos y posteriormente integrarlos en un único modelo.

Entre las principales ventajas de Xacro se encuentran:

- Reutilización de componentes.
- Organización modular del código.
- Simplificación del mantenimiento.
- Generación automática de modelos complejos.

En esta práctica se emplearon archivos Xacro para describir la estructura del robot SCARA y sus sensores, facilitando la organización del proyecto y permitiendo modificar componentes específicos sin afectar el resto del sistema.

## Robot SCARA

SCARA (Selective Compliance Assembly Robot Arm) es un tipo de manipulador serial ampliamente utilizado en aplicaciones industriales debido a su rapidez, precisión y facilidad de control. Este tipo de robot es especialmente adecuado para tareas de ensamblaje, manipulación de objetos y procesos automatizados que requieren movimientos repetitivos en un plano horizontal.

El robot utilizado en esta práctica está compuesto por tres articulaciones rotacionales y varios enlaces rígidos conectados mediante juntas de tipo revoluta. Cada articulación aporta un grado de libertad al sistema y permite modificar la posición y orientación del efector final dentro de su espacio de trabajo.

La estructura cinemática del robot fue descrita mediante archivos Xacro y posteriormente visualizada utilizando RViz2.

## RViz2

RViz2 es una herramienta de visualización tridimensional incluida dentro del ecosistema ROS 2. Su función principal consiste en representar información generada por nodos robóticos, incluyendo modelos de robots, transformaciones, sensores y trayectorias.

RViz2 permite verificar que la descripción URDF/Xacro de un robot ha sido interpretada correctamente por ROS 2. Asimismo, facilita la validación de configuraciones cinemáticas y la observación del comportamiento de un robot antes de ejecutar simulaciones más complejas.

Durante esta práctica se utilizó RViz2 para visualizar el robot SCARA, observar sus articulaciones y verificar la correcta interpretación de los archivos de descripción desarrollados.

## Descripción del archivo `scara_urdf.xacro`

Una vez analizada la estructura general del paquete `scara_description`, se procedió a estudiar el archivo principal encargado de describir el robot SCARA. Para ello se utilizó el archivo `scara_urdf.xacro`, el cual contiene la definición completa de la estructura geométrica y cinemática del manipulador.

El modelo inicia con la definición del sistema de referencia global (`world`) y una articulación fija (`world_joint`) que conecta dicho sistema con la base del robot (`base_link`). Posteriormente se definen tres enlaces principales (`link_1`, `link_2` y `link_3`) conectados mediante articulaciones de tipo revoluta.

Cada enlace incorpora información visual, de colisión e inercial. Las propiedades visuales permiten representar gráficamente el robot dentro de RViz2, mientras que las propiedades de colisión e inerciales permiten describir el comportamiento físico del sistema para futuras simulaciones.

Las articulaciones `link_1_joint`, `link_2_joint` y `link_3_joint` fueron definidas como articulaciones de tipo revoluta. Estas articulaciones proporcionan los grados de libertad necesarios para el movimiento del manipulador dentro de su espacio de trabajo.

Asimismo, el archivo define un enlace terminal denominado `P`, el cual representa el punto final del robot y puede emplearse posteriormente para tareas relacionadas con cinemática directa, cinemática inversa y planeación de trayectorias.

## Archivo de lanzamiento

Para visualizar el robot dentro de ROS 2 se utilizó el archivo `scara_display.launch.xml`. Este archivo tiene como finalidad automatizar la ejecución de todos los nodos necesarios para interpretar la descripción del robot y representarla gráficamente.

El archivo de lanzamiento ejecuta principalmente tres componentes:

* `robot_state_publisher`
* `joint_state_publisher_gui`
* `rviz2`

El nodo `robot_state_publisher` es responsable de publicar las transformaciones cinemáticas del robot a partir de la descripción contenida en el archivo Xacro. Gracias a este nodo, ROS 2 puede conocer la posición relativa de cada enlace y articulación.

Por otro lado, `joint_state_publisher_gui` proporciona una interfaz gráfica que permite modificar manualmente los valores articulares del robot. Esta herramienta resulta útil para validar la estructura cinemática y observar el comportamiento del manipulador ante diferentes configuraciones.

Finalmente, RViz2 utiliza toda esta información para representar gráficamente el robot y permitir la inspección visual de su estructura.
