# Previo

## Simuladores compatibles con ROS 2

ROS 2 es una plataforma de desarrollo robótico diseñada para facilitar la comunicación entre sensores, actuadores, algoritmos de control y sistemas de percepción. Una de sus principales ventajas es la capacidad de integrarse con distintos entornos de simulación, permitiendo validar sistemas robóticos antes de implementarlos físicamente.

Entre los simuladores más utilizados compatibles con ROS 2 se encuentran Gazebo, Ignition Gazebo y Webots. Gazebo es ampliamente utilizado debido a su capacidad para modelar fenómenos físicos complejos como gravedad, fricción, colisiones y dinámica multicuerpo. Además, permite integrar sensores virtuales tales como cámaras RGB, cámaras de profundidad, sensores LiDAR, sensores ultrasónicos e IMU.

El uso de simuladores en robótica permite reducir significativamente los costos de desarrollo y minimizar riesgos asociados a pruebas sobre hardware real. Gracias a estas herramientas es posible realizar múltiples iteraciones de diseño, depurar algoritmos y evaluar el comportamiento de los sistemas en entornos controlados antes de su implementación final.

## Restricciones de Gazebo

A pesar de las ventajas ofrecidas por Gazebo, existen diversas limitaciones inherentes a cualquier entorno de simulación. En primer lugar, la precisión física depende directamente de los modelos matemáticos implementados y de la capacidad computacional disponible en el sistema donde se ejecuta la simulación.

Los sensores simulados representan aproximaciones del comportamiento real, por lo que las mediciones obtenidas pueden diferir de aquellas registradas por dispositivos físicos. Asimismo, la dinámica de motores, transmisiones mecánicas y elementos estructurales suele simplificarse para disminuir el costo computacional.

Otro aspecto importante es que la simulación no puede representar completamente factores externos presentes en aplicaciones reales, como vibraciones, desgaste mecánico, ruido eléctrico o perturbaciones ambientales. Por esta razón, la simulación debe considerarse una etapa previa a la validación experimental en sistemas físicos reales.


# Desarrollo

## Descripción general del robot SCARA

El robot utilizado en esta práctica corresponde a un manipulador tipo SCARA (Selective Compliance Assembly Robot Arm). Este tipo de robots son ampliamente utilizados en aplicaciones industriales debido a su elevada velocidad, precisión y capacidad para realizar tareas repetitivas de ensamblaje, manipulación y posicionamiento de objetos.

La estructura del robot implementado está formada por tres articulaciones rotacionales, las cuales permiten modificar la configuración espacial del manipulador y posicionar el efector final dentro de su espacio de trabajo. Cada articulación es controlada mediante referencias de posición enviadas desde ROS 2 hacia Gazebo utilizando tópicos específicos.

La descripción geométrica y cinemática del robot se encuentra implementada mediante archivos Xacro. Xacro constituye una extensión de URDF que permite reutilizar código, definir propiedades y simplificar la generación de modelos complejos. Gracias a esta estructura modular es posible modificar componentes individuales del robot sin afectar el resto del sistema.

## Organización de los archivos Xacro

El archivo principal empleado para la simulación es `gz2_scara.xacro`. Este archivo actúa como punto de integración de todos los elementos que conforman el robot y el entorno de simulación.

Dentro de esta estructura se incluye `scara_urdf.xacro`, encargado de definir los enlaces, articulaciones y propiedades geométricas del manipulador. Asimismo, se incorpora `scara_gz2.xacro`, el cual contiene los plugins y controladores necesarios para que Gazebo pueda simular y controlar las articulaciones del robot.

Adicionalmente, se incluyen los archivos `camera_sensor.xacro`, `lidar_sensor.xacro`, `sonar_sensor.xacro` e `imu_sensor.xacro`, los cuales permiten agregar sensores virtuales al modelo robótico. Estos sensores pueden utilizarse para tareas de percepción, navegación, localización y control.

## Paquete scara_bringup

El paquete `scara_bringup` contiene todos los elementos necesarios para iniciar la simulación. Dentro de este paquete se encuentran los archivos launch, configuraciones de RViz, archivos de configuración del bridge y los mundos utilizados por Gazebo.

El archivo principal `gz2_scara.launch.py` es responsable de inicializar simultáneamente Gazebo, RViz2, el robot_state_publisher y ros_gz_bridge. Gracias a esta configuración es posible visualizar el robot, controlar sus articulaciones y monitorear los datos generados durante la simulación.

## Comunicación entre ROS 2 y Gazebo

La comunicación entre ROS 2 y Gazebo se realiza mediante el paquete `ros_gz_bridge`. Este componente funciona como un puente de comunicación capaz de traducir mensajes entre ambos entornos.

Gracias a este mecanismo es posible enviar referencias de posición desde ROS 2 hacia Gazebo mediante mensajes tipo Float64 y recibir información proveniente de sensores y estados articulares. Durante la ejecución de la práctica se verificó el correcto funcionamiento del bridge observando la transmisión de mensajes hacia los controladores de las articulaciones del robot.

## Planeación de movimiento mediante q_plan.py

Para generar movimiento automático del robot se empleó el programa `q_plan.py`. Este nodo implementa una estrategia de planeación basada en cinemática inversa y trayectorias suaves definidas mediante polinomios de quinto orden.

Inicialmente se define una posición de origen y una posición destino dentro del espacio de trabajo del manipulador. Posteriormente se calcula la cinemática inversa correspondiente para obtener los valores articulares requeridos. Finalmente se genera una trayectoria continua que minimiza cambios bruscos de velocidad y aceleración, enviando referencias articulares a través de los tópicos `/joint1/cmd_pos`, `/joint2/cmd_pos` y `/joint3/cmd_pos`.


## Paquete scara_bringup

El paquete `scara_bringup` contiene los archivos necesarios para iniciar la simulación.

El archivo principal `gz2_scara.launch.py` ejecuta simultáneamente:

- Gazebo
- RViz2
- robot_state_publisher
- ros_gz_bridge

permitiendo visualizar y controlar el robot desde ROS 2.


## Comunicación Gazebo - ROS

La comunicación entre Gazebo y ROS 2 se realiza mediante el paquete `ros_gz_bridge`.

Este bridge permite intercambiar información entre ambos entornos, incluyendo estados articulares, sensores y comandos de control. Gracias a este mecanismo es posible enviar referencias de posición desde ROS hacia Gazebo y visualizar el comportamiento del robot en tiempo real.
